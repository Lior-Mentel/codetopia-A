```python
#Third graph: HeatMap

def plot_heatmap(data): #Function that create the HeatMap
    # Choosing from the data the relevent columns we want to show in the HeatMap
    age_politics_df = data[['Age from file', 'In Israeli politics it is used to talk about left and right. Where would you place yourself on a scale where 1 is left, 4 is center and 7 is right?']]

    # Rename columns for easier access
    age_politics_df.columns = ['Age', 'Political_Orientation']

    # Convert 'Political_Orientation' column to numeric (So that we can insert the values â€‹â€‹into the heat map and analyze the data accordingly), coercing errors
    age_politics_df['Political_Orientation'] = pd.to_numeric(age_politics_df['Political_Orientation'], errors='coerce')

    # Drop rows with NaN values in 'Political_Orientation'
    age_politics_df.dropna(subset=['Political_Orientation'], inplace=True)

    # Create a pivot table for the heat map
    # Use pivot tables for summarizing and aggregating the data,and in that way we will find patterns or relationships in the data.
    heatmap_data = age_politics_df.pivot_table(index='Age', columns='Political_Orientation', aggfunc='size', fill_value=0)

    # Plot the HeatMap
    plt.figure(figsize=(12, 8))
    sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', cbar=True, linewidths=.5)
    plt.title('Political Orientation by Age Group')  #Title of the graph
    plt.xlabel('Political Orientation (1 = Left, 4 = Center, 7 = Right)')  #Label for axis x
    plt.ylabel('Age Group')  #Label for zxis y
    plt.show()  #Showing the graph

plot_heatmap(democratic_survey) #Activate the function on our data
```


We can see in this HeatMap graph that in all age gruops we have more people who voted to center and right parties, mostly in the age groups 25-34 and 35-44 there is a majority of voters mainly to the right.

That is, there is a high correlation between these age groups and votes with a value of 4 or higher,It means these age groups mainly vote for the center and the right.

we can see that in gruops age 55-64 and 65+ there is no unequivocal answer for the votes but also have leaning to the right.

**We can see in the graph in general that there is less concentration of people who voted for the left parties in all ages gruop.**

```python
#Creating a function that checks whether the value in a cell is a string,
#if so separates the number and the string and puts the number in a new column that we added to the data and saves it.

def split_if_string(value):
    if isinstance(value, str):
        parts = value.split('-')  #split the value by a "-"
        if len(parts) > 1:
            political_place = parts[0]
            return political_place
    return value
```
